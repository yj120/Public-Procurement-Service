# 항목별 배정예산

```python
test01 = new_service[new_service['업종제한여부'] == 'Y']
test02 = new_service[new_service['업종제한여부'] == 'N']
```

```python
test01_groupby = test01.groupby(by=['수요기관코드'], as_index=False).sum()
test02_groupby = test02.groupby(by=['수요기관코드'], as_index=False).sum()
```

```python
test01_total=test01_groupby.배정예산.sum()


test02_total=test02_groupby.배정예산.sum()


test_set = [test01_total,test02_total]


test01_count = test01_groupby.수요기관코드.count()
test02_count = test02_groupby.수요기관코드.count()

print("Y:%d"%test01_count,"N:%d"%test02_count)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.bar(range(len(test_set)), test_set,color=['dodgerblue','dodgerblue'])
x=np.arange(2)
section = ['Y','N']
plt.xticks(x,section)
plt.ylabel("배정예산")
plt.xlabel("업종제한 구분")
plt.title("업종제한 유무에 따른 배정예산 크기비교",fontsize=15)
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.show()
```

```python
service_contest.loc[service_contest['입찰공고차수'] >= 1, '입찰공고차수유무'] = 'y'
service_contest.loc[service_contest['입찰공고차수'] == 0, '입찰공고차수유무'] = 'n'

service_contest.columns
```

# 용역구분별로 따져보기

```python
service_contest.용역구분.value_counts()
```

```python
service_ordinary = service_contest[service_contest['용역구분'] == '일반용역']
service_tech = service_contest[service_contest['용역구분'] == '기술용역']
service_null = service_contest[service_contest['용역구분'].isnull()]

col_name = ["수요기관코드","입찰공고차수","배정예산","입찰공고차수유무"]


service_ordinary_col_down = pd.DataFrame(data=service_ordinary, columns=col_name)
service_tech_col_down = pd.DataFrame(data=service_tech, columns=col_name)
service_null_col_down = pd.DataFrame(data=service_null, columns=col_name)

ser_o_sum = service_ordinary_col_down.배정예산.sum()
ser_t_sum = service_tech_col_down.배정예산.sum()
ser_n_sum = service_null_col_down.배정예산.sum()

total_sum = [ser_o_sum,ser_t_sum,ser_n_sum]
```

```python
section = ["일반용역","기술용역","미분류"]
section_num = np.arange(len(total_sum))
values = total_sum

plt.bar(section_num, values, color=['C2','C2','C2'])
plt.xticks(section_num, section)
plt.rcParams["figure.figsize"] = (10,6)
plt.xticks(section_num,section)
plt.ylabel("배정예산합계")
plt.xlabel("용역구분에 따른 배정예산 크기")
plt.title("용역구분에 따른 배정예산 크기비교",fontsize=20)
plt.rcParams['font.family'] = 'Malgun Gothic'

plt.show()
```
