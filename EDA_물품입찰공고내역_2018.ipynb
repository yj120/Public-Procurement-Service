# 2. 데이터 전처리 (분할)

- sklearn

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
```

```python
# 학습데이터랑 테스트데이터랑 랜덤하게 추출해서 분할, 특정조건 하에 랜덤하게 뽑는데
# random_state를 맞추어 주게 되면 어떤 컴퓨터에서도 똑같이 분할함
X_train, X_test, Y_train, Y_test =train_test_split(X,Y,test_size=0.3, 
                                                   random_state=2020) # 학습데이터, 테스트데이터 분류

print(X_train.shape) #row, col 
print(X_test.shape)
print(Y_train.shape) # row 갯수
print(Y_test.shape)
```

# 3. 학습

```python
model = DecisionTreeClassifier()
model.fit(X_train,Y_train)
```

# 4. 평가

```python
# 학습이 잘 되었는지 평가
# 예측한다.

Y_train_pred = model.predict(X_train) # 학습데이터로 예측 # 알고리즘이 학습을 얼마나 잘 시켰는가?
Y_test_pred = model.predict(X_test) # 학습되지않은 데이터로 예측 # 새로운 데이터가 들어올 때, 얼마나 좋은 성능을 나타내는가?
```

```python
print(classification_report(Y_train,Y_train_pred)) # 학습에 참여한 데이터
```

```python
print(classification_report(Y_test,Y_test_pred)) # 학습에 참여하지 않은 데이터
```

# KNN (K Nearest Neighbors): K 최근접 이웃 알고리즘

#### 새로운 데이터의 분류 결과가 K 개 최근접 이웃의 클래스에 의해서 결정되며, 데이터는 가장 많이 할당되는 클래스로 분류하게 된다.



```python
from sklearn.neighbors import KNeighborsClassifier
# 1. 모델 선언
knn = KNeighborsClassifier()
# 2. 모델 학습
knn.fit(X_train,Y_train)
```

```python
# 3. 예측
prediction = knn.predict(X_test)
```

```python
# 4. 평가
m=(prediction == Y_test).mean()
print("평균정확도:",m)
```

```python
#n_neighnors를 9개로 설정하여 다시 예측해본다:

knn2 = KNeighborsClassifier(n_neighbors=4)
knn2.fit(X_train,Y_train)
knn2_pred = knn2.predict(X_test)
```

```python
xgbr.fit(X_train,Y_train)
score= xgbr.score(X_train,Y_train)
print("Training score: ",score) # -> 입찰공고차수	합계수량	합계금액	추정가격	배정예산	낙찰하한율
```
